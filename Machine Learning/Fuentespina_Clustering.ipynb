### **Clustering using sklearn (with pandas and pyplot)**

<font color="red">File access required:</font> In Colab this notebook requires first uploading files **Cities.csv** and **Players.csv** using the *Files* feature in the left toolbar. If running the notebook on a local computer, simply ensure these files are in the same workspace as the notebook.

```python
# Set-up
import csv
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import cluster
```

```python
# Read Cities.csv into dataframe
f = open('Cities.csv')
cities = pd.read_csv(f)
```

```python
# K-means clustering based on longitude-latitude
kmeans = cluster.KMeans(5)
kmeans.fit(cities[['longitude', 'latitude']])
labels = kmeans.labels_
plt.scatter(cities['longitude'], cities['latitude'], c=labels)
plt.show()
# Try different values for K
```

```python
# Now with centroids
kmeans = cluster.KMeans(5)
kmeans.fit(cities[['longitude', 'latitude']])
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
plt.scatter(cities['longitude'], cities['latitude'], c=labels)
plt.scatter(centroids[:,0], centroids[:,1], marker='x', c='black')
plt.show()
```

```python
# K-means clustering based on temperature
kmeans = cluster.KMeans(3)
kmeans.fit(cities[['temperature']])
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
plt.scatter(cities['longitude'], cities['latitude'], c=labels)
plt.show()
print(centroids)
```

### <font color="green">**Your Turn: Clustering on World Cup Data**</font>

```python
# Read Players.csv into dataframe
f = open('Players.csv')
players = pd.read_csv(f)
list(players.columns)
```

```python
# Create each scatterplot for each numerical columns
cols = ['minutes', 'shots', 'passes', 'tackles', 'saves']

for i in range(len(cols)):
    for j in range(i + 1, len(cols)):
        x = cols[i]
        y = cols[j]
        plt.scatter(players[x], players[y])
        plt.xlabel(x.capitalize())
        plt.ylabel(y.capitalize())
        plt.title(f'{x.capitalize()} vs {y.capitalize()}')
        plt.show()
```

```python
# Create a clustering from the Players data and show it
# visually in a scatterplot. No need for centroids.

features = ['minutes', 'shots']

kmeans = cluster.KMeans(4)
kmeans.fit(players[features])

labels = kmeans.labels_

plt.scatter(players['minutes'], players['shots'], c=labels)
plt.xlabel('Minutes')
plt.ylabel('Shots')
plt.title('Player Clustering')
plt.show()
```
